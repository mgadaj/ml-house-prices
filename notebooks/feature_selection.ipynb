```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

## Load train and test sets

```python
pd.pandas.set_option('display.max_columns', None)

X_train = pd.read_csv('../data/processed/train_processed.csv')
X_test = pd.read_csv('../data/processed/test_processed.csv')

X_train.head()
```

## Select the target from the sets

```python
y_train = X_train['SalePrice']
y_test = X_test['SalePrice']
```

## Drop unnecessary variables

```python
X_train.drop(['Id', 'SalePrice'], axis=1, inplace=True)
X_test.drop(['Id', 'SalePrice'], axis=1, inplace=True)
```

## Lasso regression for feature selection

```python
selector_lr = SelectFromModel(Lasso(alpha=0.005, random_state=0))
selector_lr.fit(X_train, y_train)
```

```python
selected_features = X_train.columns[(selector_lr.get_support())]

print(f'Number of features: {X_train.shape[1]}')
print(f'Selected features: {len(selected_features)}')
```
